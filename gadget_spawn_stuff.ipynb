```python
from common import *
spawns = load_json('src/main/resources/defaults/data/GadgetSpawns.json')
group_ids = {d.get('groupId', None) for d in spawns} - {None}
dir = os.path.join(GC_PATH, 'resources/Scripts')
query_str = r'\b[12][03-7]\d{7}\b'
query = re.compile(query_str)
potential_gid_hits = !grep -n -P "{query_str}" -r {dir}
def split_grep(input):
    # s = input[len(dir)+1:].split(':')
    # return (':'.join(s[:2]), ':'.join(s[2:]))
    filename, _, line = input[len(dir)+1:].partition(':')
    return (filename, line)
potential_gid_hits = [split_grep(line) for line in potential_gid_hits]

gid_hits = {}
for (filename, line) in potential_gid_hits:
    for match in query.findall(line):
        m = int(match)
        if m in group_ids:
            d = gid_hits.get(m, {})
            d[filename] = d.get(filename, []) + [line]
            gid_hits[m] = d
```

```python
simplex_hits = []
complex_hits = []

def is_simplex(gid: int, files: dict) -> bool:
    if len(files) > 2:
        return False
    for file, hits in files.items():
        if str(gid) in file:
            continue
        if len(hits) > 1:
            return False
    return True

for gid, files in gid_hits.items():
    if is_simplex(gid, files):
        simplex_hits.append(gid)
    else:
        complex_hits.append(str(gid))
        for file, hits in files.items():
            complex_hits.append(f'\t{file}: {hits}')
print(f'Simple hits: {len(simplex_hits)}')
print(', '.join([str(i) for i in simplex_hits]))
print(f'\n\nComplex hits: {len(complex_hits)}')
print('\n'.join(complex_hits))
```
